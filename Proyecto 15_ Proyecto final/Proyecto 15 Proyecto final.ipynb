## Targets

```python
def create_relative_target(df, target_window):
  target_name = f"target_{target_window}d"
  target_df = df.copy()

  target_df[target_name] = target_df["PX_LAST"].shift(-target_window)
  return target_df
```

```python
df_target = create_relative_target(df, 30)
df_target.head()
```

## Features

Ahora que contamos con una columna completa de valores, continuaremos a crear los features y targets para nuestos modelos. Nos serviremos de la siguiente función para hacerlo.

```python
def make_features(data, max_lag, rolling_mean_size):

  #data['year'] = data.index.year
  #data['month'] = data.index.month
  #data['day'] = data.index.day
  #data['dayofweek'] = data.index.dayofweek
  #features = data.copy()

  for lag in range(1, max_lag + 1):
    data['lag_{}'.format(lag)] = data['PX_LAST'].shift(lag)

  data['rolling_mean_custom'] = data['PX_LAST'].rolling(rolling_mean_size).mean()
  data['rolling_mean_20'] = data['PX_LAST'].rolling(20).mean()
  data['rolling_mean_50'] = data['PX_LAST'].rolling(50).mean()
  data['rolling_mean_200'] = data['PX_LAST'].rolling(200).mean()
  return data
```

# Análisis de predicciones

Ahora que contamos con todos los modelos entrenados con los datos hasta el año 2021, realizaremos las predicciones para el año 2022. Para esto definiremos 2 funciones que nos permitirán crear los features y targets especificos para cada empresa.

```python
def feat_tar(df):
  df_info = df.loc[:,['PX_LAST', 'BEST_TARGET_PRICE',	'BEST_ANALYST_RATING',	'TOT_BUY_REC',	'TOT_HOLD_REC',	'TOT_SELL_REC']]
  df_info = df_info.ffill(axis=0)
  df_tar = create_relative_target(df_info, 30)
  df_ft = make_features(df_tar, 5, 5)
  df_ft.dropna(inplace=True)
  valid = df_ft.query(f"(index >= '2022')")
  valid_feat = valid.drop(['PX_LAST', 'target_30d'], axis=1)
  valid_tar_d = valid['PX_LAST']
  valid_tar_30d = valid['target_30d']
  return valid_feat, valid_tar_d, valid_tar_30d

def predictions (features):
  pred_d = reg_d.predict(features)
  pred_30d = reg_30d.predict(features)
  pred_60d = reg_60d.predict(features)
  pred_90d = reg_90d.predict(features)
  return pred_d, pred_30d, pred_60d, pred_90d
```

Ahora que tenemos las funciones listas, comenzaremos a trabajar con las predicciones de las empresas. Para esto hemos preparado el codigo para revisar todas las empresas o un conjunto de empresas de diferentes sectores económicos.

```python
#stocks = ['GOOGL', 'SHEL', 'AIG', 'UPS', 'MSFT', 'IBE', 'MRK']
stocks = company_stock_df['company'].unique()

for stock in stocks:
  company = company_stock_df[company_stock_df['company'] == stock]
  comp_feat, com_tar, comp_tar_30 = feat_tar(company)
  pred_d, pred_30d, pred_60d, pred_90d = predictions(comp_feat)
  com_tar.reset_index(drop=True, inplace=True)

  plt.figure(figsize=(27, 5))
  sns.lineplot(data=com_tar, label=stock)
  sns.lineplot(data=pred_d, color="red", ls="--", label="prediction 1 day")
  sns.lineplot(data=pred_30d, color="orange", ls="--", label="prediction 30 days")
  sns.lineplot(data=pred_60d, color="yellow", ls="--", label="prediction 60 days")
  sns.lineplot(data=pred_90d, color="green", ls="--", label="prediction 90 days")
```
