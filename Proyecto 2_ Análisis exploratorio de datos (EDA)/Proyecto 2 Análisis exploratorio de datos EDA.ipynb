## Corregir los tipos de datos


### Columna `condition`

Queremos cambiar las categorías de `condition` por valores númericos, para esto crearemos una nueva columna `condition_id` con los valores necesarios.

Para cada valor de la columna `condition` haremos el siguiente cambio:
- new = 5
- like new = 4
- excellent = 3
- good = 2
- fair = 1
- salvage = 0

```python
df.groupby('condition')['model'].count()

# Veamos que valores tenemos y cuanto se repite cada uno
```

```python
def condition_id_fill (row):
    if row['condition'] == 'new':
        return 5
    if row['condition'] == 'like new':
        return 4
    if row['condition'] == 'excellent':
        return 3
    if row['condition'] == 'good':
        return 2
    if row['condition'] == 'fair':
        return 1
    if row['condition'] == 'salvage':
        return 0
```

```python
df['condition_id'] = df.apply(condition_id_fill, axis=1)
len(df['condition_id'])

# Veamos si tenemos todas las filas llenadas
```

```python
df.groupby('condition_id')['model'].count()

# Veamos si la nueva columna tiene la misma distribución que la original
```

### Columnas `cylinders` y `model_year`

Ya que estas dos columnas necesitan el mismo arreglo, lo haremos en conjunto. Para ambas, necesitamos cambiar los  tipos de datos componen estas columnas. Ahora mismo el tipo de datos es `float`, es decir, números con decimales, pero estas dos medidas son siempre enteros. Entonces, transformaremos los valores de estas columnas en datos tipo `int`.

```python
# Creamos una función que facilite nuestro trabajo
def int_converter (value):
    return int(value)

# Aplicamos la función a ambas columnas
df['cylinders'] = df['cylinders'].apply(int_converter)
df['model_year'] = df['model_year'].apply(int_converter)
```

### Columna `is_4wd`

Transformaremos esta columna a valores booleanos, pues solo necesitamos saber si tiene o no este tipo de tracción. Así nos evitaremos una explicación y descripción de los datos al compartirlos, facilitando su entendimiento. Además, podremos hacer todos los análisis matemáticos y estadísticos que podiamos hacer con sus valores numéricos.

```python
# Creamos la función que nos ayudará a convertir los datos
def bool_converter (value):
    if value==1:
        return True
    else:
        return False
    
# Aplicamos la función a la columna
df['is_4wd'] = df['is_4wd'].apply(bool_converter)

# Revisemos nuestro trabajo en esta sección
df.info()
```

## Enriquecer datos

```python
# Agrega valores de fecha y hora para cuando se colocó el anuncio

# Convertimos el `str` de `date_posted` al tipo `datetime64`, de esta forma obtenemos la fecha en la que se publica el vehículo
# Por desgracia, se nos hace imposible encontrar la hora a la que se sube

df['date_posted'] = pd.to_datetime(df['date_posted'], format='%Y-%m-%d')
df.info()
```

```python
# Agrega los años del vehículo cuando el anuncio se colocó

def car_age (row):
    return row['date_posted'].year - row['model_year'] + 1

df['car_age'] = df.apply(car_age, axis=1)
df.head()
```

```python
# Agrega el millaje promedio del vehículo por año

df['km_avg'] = df['odometer'] / df['car_age']
df.head()
```

## Estudiar parámetros principales sin valores atípicos

Comencemos esta sección definiendo algunas funciones que necesitaremos para nuestro trabajo.

```python
# Esta función nos permite encontrar el valor del bigote superior y la usaremos en la siguiente función
def iqr_finder (data, column):
    return data[column].quantile(0.75) + (data[column].quantile(0.75) - data[column].quantile(0.25))*1.5

# Esta función nos permitirá crear una serie con los valores típicos que se encuentren bajo el límite superior del bigote mayor
def iqr_limit (data, column):
    iqr = iqr_finder(data, column)
    data_set = data[column]
    data_set = data_set[data_set<=iqr]
    return data_set

# Esta función es muy parecida a la anterior, pero nos permite encontrar los valores fuera del límite superior del bigote mayor
def iqr_off_limit (data, column):
    iqr = iqr_finder(data, column)
    data_set = data[column]
    data_set = data_set[data_set>iqr]
    return data_set
```
