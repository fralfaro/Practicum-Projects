### Columna `user_score`

```python
df['user_score'].describe()
```

```python
df.boxplot(column='user_score')
```

Dentro de esta columna, la media y mediana parecen ser bastante cercanas, pero no lo suficente. Al ver el gráfico de caja encontramos la razón, existen muchos valores por debajo del bigote menor. Estos valores afectan a nuestras medidas de posición. Por ser un grupo de datos asimétrico y con valores atípicos, se utilizará la mediana para llenar los datos vacíos.

```python
# Creamos la tabla que relaciona las variables y usa la medida de posición escogida
df_user_scores = df.pivot_table(index=['platform', 'year_of_release'], columns='genre', values='user_score', aggfunc='median')
df_user_scores.head(10)
```

```python
# Creamos la función que aplicaremos
def fill_user(genre, platform, year_of_release):
    try:
        return df_user_scores[genre][platform][year_of_release]
    except:
        return 'error'
```

```python
df_fill_user = df.apply(lambda row: fill_user(row['genre'], row['platform'], row['year_of_release']), axis=1)
df['user_score'] = df['user_score'].fillna(df_fill_user)
```

```python
# Comprobemos cuantos valores vacíos nos quedan
len(df[df['user_score'].isna() == True])
```

```python
# Comprobemos cuantos errores tuvimos
len(df[df['user_score']=='error'])
```

Con estos pasos hemos logrado llenar casi 7 mil valores vacíos que teníamos en esta columna, obteniendo casi el doble de datos conocidos. Se volverán a repetir los pasos de manera resumida para lidiar con los 2166 datos que nos faltan aún.

```python
# Vacíamos los valores 'error' en la columna
df['user_score'] = df['user_score'].replace('error', np.nan)

# Creamos la tabla con menos valores relacionados
df_user_scores_2 = df.pivot_table(index='platform', columns='genre', values='user_score', aggfunc='median')

# Creamos la nueva función que aplicaremos
def fill_user_2(genre, platform):
    try:
        return df_user_scores_2[genre][platform]
    except:
        return 'error'
    
# Creamos la lista de donde sacaremos los valores vacíos para rellenar
df_fill_user_2 = df.apply(lambda row: fill_user_2(row['genre'], row['platform']), axis=1)

# Rellenamos los valores vacíos
df['user_score'] = df['user_score'].fillna(df_fill_user_2)

# Comprobamos los resultados de nuestro trabajo
print('Valores vacíos:')
print(len(df[df['user_score'].isna() == True]))
print()
print('Valores error:')
print(len(df[df['user_score']=='error']))
```

Con los nuevos cambios, nos quedan solo un poco más de mil datos vacíos. Puede parecer un número alto, pero podemos tener una mejor perspectiva al compararlo con los más de 9 mil datos vacíos que teníamos originalmente.
```python
# Primero hagamos un listado de los años
year_list = list(df_by_year.columns.values)

# Ahora creamos la función que nos permitirá contar los años de vida de cada plataforma
def platform_life (row):
    years = []
    counter = 0
    for value in row:
        if pd.isna(value)==False:
            years.append(year_list[counter])
        counter += 1
    return (max(years)-min(years)+1)
    # Debemos sumar 1 para considerar el primer año de la plataforma

# Ahora apliquemos la función
df_by_year['platform_life'] = df_by_year.apply(platform_life, axis=1)
```

# Comprobación de hipótesis

En esta sección, comprobaremos las hipótesis. Para esto consideraremos un alpha de 5%. Como podemos considerar que cada plataform y género de juego se comporta de manera diferente y tendrá cantidades de datos diferentes, utilizaremos el método `ttest_ind()`.

## Hipótesis 1
